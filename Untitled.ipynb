```python
import pandas as pd
import altair as alt


sheet_id = '1sREWbJdEYFjckWHcgAaUTnb6uQIuyM-bs-Un12ZmVaA'
gid = '598429315'
url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&gid={gid}'
df = pd.read_csv(url)


def clean_column(df: pd.DataFrame, col: str) -> pd.DataFrame:
    """Coerce column to numeric and drop NaN rows."""
    df = df.copy()
    df[col] = pd.to_numeric(df[col], errors='coerce')
    return df.dropna(subset=[col])


def aggregate(df: pd.DataFrame, col: str) -> pd.DataFrame:
    """Aggregate by country + year, summing the target column."""
    return df.groupby(['countryname', 'year'])[col].sum().reset_index()


def pct_change_from_first(df: pd.DataFrame, col: str) -> pd.DataFrame:
    """Add a % change column relative to each country's first year."""
    df = df.copy()
    first = df.groupby('countryname')[col].transform('first')
    df[f'{col}_pct'] = ((df[col] - first) / first) * 100
    return df


def facet_chart(data: pd.DataFrame, col: str, y_title: str,
                title: str, pct: bool = False) -> alt.FacetChart:
    """
    Build a steelblue line + dot facet chart.

    Parameters
    ----------
    data    : aggregated DataFrame
    col     : column to plot on Y axis
    y_title : Y-axis label
    title   : top-level chart title
    pct     : if True, format Y labels as percentages
    """
    axis_kwargs = dict(tickCount=3, grid=False)
    if pct:
        axis_kwargs['labelExpr'] = "datum.value + '%'"

    encode = dict(
        x=alt.X('year:O', title=None),
        y=alt.Y(f'{col}:Q', title=y_title, axis=alt.Axis(**axis_kwargs)),
    )

    line = alt.Chart(data).mark_line(color='steelblue').encode(**encode)
    dots = alt.Chart(data).mark_point(filled=True, size=50, color='steelblue').encode(**encode)

    return (
        (line + dots)
        .properties(title=title)
        .facet(
            facet=alt.Facet(
                'countryname:N',
                header=alt.Header(titleOrient='bottom', labelOrient='bottom'),
                title='Country Name',
            ),
            columns=4,
        )
        .resolve_scale(y='independent')
        .resolve_axis(x='independent')
        .properties(
            title=alt.TitleParams(
                text=title, fontSize=16, fontWeight='bold', anchor='middle'
            )
        )
    )


METRICS = [
    ('c1_visits',     'Number of Visits',        'SP Website Visits Over Years by Country (log scale)'),
    ('c2_user',       'Number of Users',          'Trends of C2 Users Over Years by Country'),
    ('c3_pdoDeliver', 'Number of Deliveries',     'Trends of C3 PDO Deliveries Over Years by Country'),
    ('c4_events',     'Number of Events',         'Trends of C4 Events Over Years by Country'),
    ('c6_eAttendees', 'Number of Attendees',      'Trends of C6 Event Attendees Over Years by Country'),
    ('c8_allEvent',   'Number of All Events',     'Trends of C8 All Events Over Years by Country'),
    ('c10_pdoStored', 'Number of PDO Stored',     'Trends of C10 PDO Stored Over Years by Country'),
    ('c13_staff',     'Number of Staff',          'Trends of C13 Staff Over Years by Country'),
    ('c14_nfunds',    'National Funds (€)',       'Trends of C14 National Funds Over Years by Country'),
    ('c15_cstaff',    'Number of Contract Staff', 'Trends of C15 Contract Staff Over Years by Country'),
    ('c16_cfunds',    'Contract Funds (€)',       'Trends of C16 Contract Funds Over Years by Country'),
    ('c19_pub',       'Number of Publications',   'Trends of C19 Publications Over Years by Country'),
]


charts = {}

for col, y_title, title in METRICS:
    clean_df = clean_column(df, col)
    agg_df = aggregate(clean_df, col)
    charts[col] = facet_chart(agg_df, col, y_title, title)


clean_visits = clean_column(df, 'c1_visits')
pct_df = pct_change_from_first(aggregate(clean_visits, 'c1_visits'), 'c1_visits')

chart_pct = facet_chart(
    pct_df, 'c1_visits_pct',
    y_title='% Change from First Year',
    title='C1 Visits % Change from First Year by Country',
    pct=True,
)


for col, _, _ in METRICS:
    charts[col].display()

chart_pct.display()
```
